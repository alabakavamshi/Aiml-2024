import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset into a dataframe
url = 'C:\Users\vamsh\Downloads\titanic_train.csv'
df = pd.read_csv(url)

# Check data types of each feature
print("Data Types of Features:")
print(df.dtypes)

# Generate summary statistics
summary = df.describe(percentiles=[.25, .50, .75, .90])
print("\nSummary Statistics:")
print(summary)

print("\nColumn Names:")
print(df.columns.tolist())

numeric_features = ['Age', 'SibSp', 'Parch', 'Fare']
df[numeric_features].hist(bins=20, figsize=(12, 8))
plt.suptitle("Numeric Features Distribution")
plt.show()

# Check relative size of survived/un-survived
sns.countplot(data=df, x='Survived')
plt.title("Survived vs Not Survived")
plt.show()

# Check patterns on gender
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title("Survival Count by Gender")
plt.show()

# Passenger class-wise survival rate
sns.barplot(data=df, x='Pclass', y='Survived')
plt.title("Passenger Class-wise Survival Rate")
plt.show()

# Siblings count distribution
sns.countplot(data=df, x='SibSp', hue='Survived')
plt.title("Siblings Count Distribution and Survival")
plt.show()

# Overall age distribution
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title("Overall Age Distribution")
plt.show()

# Class-wise age distribution
sns.boxplot(data=df, x='Pclass', y='Age')
plt.title("Class-wise Age Distribution")
plt.show()

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset into a dataframe
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Check data types of each feature
print("Data Types of Features:")
print(df.dtypes)

# Generate summary statistics
summary = df.describe(percentiles=[.25, .50, .75, .90])
print("\nSummary Statistics:")
print(summary)

# List the names of columns/features
print("\nColumn Names:")
print(df.columns.tolist())

# Perform exploratory analysis
# Plot numeric features
numeric_features = ['Age', 'SibSp', 'Parch', 'Fare']
df[numeric_features].hist(bins=20, figsize=(12, 8))
plt.suptitle("Numeric Features Distribution")
plt.show()

# Check relative size of survived/un-survived
sns.countplot(data=df, x='Survived')
plt.title("Survived vs Not Survived")
plt.show()

# Check patterns on gender
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title("Survival Count by Gender")
plt.show()

# Passenger class-wise survival rate
sns.barplot(data=df, x='Pclass', y='Survived')
plt.title("Passenger Class-wise Survival Rate")
plt.show()

# Siblings count distribution
sns.countplot(data=df, x='SibSp', hue='Survived')
plt.title("Siblings Count Distribution and Survival")
plt.show()

# Overall age distribution
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title("Overall Age Distribution")
plt.show()

# Class-wise age distribution
sns.boxplot(data=df, x='Pclass', y='Age')
plt.title("Class-wise Age Distribution")
plt.show()
```
