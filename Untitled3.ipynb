```python
import matplotlib.pyplot as plt
people = ['vam','krishna','sad','dsd','fg','fgdgfd','ff','fh']
age =[25, 30, 35, 40, 45,43,43,65]
height =[145, 151, 165, 173,178,169,156,190]
weight=[45, 55, 65, 75,78,89,76,55]
```

```python
plt.scatter(weight,height)
plt.title("Relationship between weight and height of patients")
plt.ylabel("Height in meters")
plt.xlabel("Weight in kgs")
plt.show()
```

```python
plt.scatter(age,height)
plt.title("Relationship between weight and height of patients")
plt.ylabel("Height in meters")
plt.xlabel("Weight in kgs")
plt.show()
```

```python
plt.scatter(age,weight)
plt.title("Relationship between weight and height of patients")
plt.ylabel("Height in meters")
plt.xlabel("Weight in kgs")
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.title("Peoples weight in kgs",fontsize=18,fontstyle='italic')
plt.ylabel("Weight  in kgs")
plt.xlabel("People")
plt.bar(people,weight)
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.title("Peoples weight in kgs",fontsize=18,fontstyle='italic')

plt.xlabel("weight's in kgs")
plt.hist(weight)
plt.show()
```
